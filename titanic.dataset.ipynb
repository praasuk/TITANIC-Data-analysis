# Investigate whether there were any outlier passengers who paid significantly more or less than average in each class, and how they fared in survival.

```python
df.head(2)
```

```python
q_w = df.groupby('Pclass')['Fare'].mean()
q_w
```

```python
def find_outliers(group):
    Q1 = group['Fare'].quantile(0.25)
    Q3 = group['Fare'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group['Fare'] < lower_bound) | (group['Fare'] > upper_bound)]

# Apply within each Pclass group
fare_outliers = df.groupby('Pclass', group_keys=False).apply(find_outliers)

print(fare_outliers[['Pclass', 'Fare', 'Name']])
```

```python
# Check if 'Survived' is already in fare_outliers
if 'Survived' not in fare_outliers.columns:
    outliers_with_survival = pd.merge(fare_outliers, df[['PassengerId', 'Survived']], on='PassengerId', how='left')
else:
    outliers_with_survival = fare_outliers.copy()

# Now group by Pclass and calculate survival rate
survival_stats = outliers_with_survival.groupby('Pclass')['Survived'].agg(
    total_passengers='count',
    survival_rate='mean'
).reset_index()

# Optional: Convert survival rate to percentage
survival_stats['survival_rate'] = survival_stats['survival_rate'].round(2) * 100

print(survival_stats)
```

# Do passengers with the same last_name but different sir name values indicate non-family relationships (e.g., staff or unrelated)? How does their survival pattern compare to actual family groups?



```python
df.head(2)
```

```python
# 1. Extract last name and title ("sir name")
df['LastName'] = df['Name'].str.split(',').str[0].str.strip()
df['Title'] = df['Name'].str.extract(r',\s*([^\.]*)\.', expand=False).str.strip()

# 2. Define common family-related titles
family_titles = ['Mr', 'Mrs', 'Miss', 'Master']

# 3. Count unique titles per last name
title_counts = df.groupby('LastName')['Title'].nunique()

# 4. Identify:
# - Likely families: last names with 1 title or only family-like titles
# - Possibly unrelated: last names with 2+ unique titles and at least one uncommon one

# Create flag columns
df['IsFamilyTitle'] = df['Title'].isin(family_titles)
df['GroupType'] = 'UnrelatedSameLastName'  # default

# Mark as family if all titles under same last name are family-like
def label_family(group):
    unique_titles = group['Title'].unique()
    if all(title in family_titles for title in unique_titles):
        group['GroupType'] = 'LikelyFamily'
    return group

df = df.groupby('LastName').apply(label_family)

# 5. Compare survival rates
group_survival = df.groupby('GroupType')['Survived'].agg(
    total_members='count',
    survival_rate='mean'
).reset_index()

group_survival['survival_rate'] = (group_survival['survival_rate'] * 100).round(2)

# 6. Show results
print(group_survival)
```

```python
# Example: Separate tickets into numeric and alphanumeric

# Create boolean masks
is_numeric_ticket = df['Ticket'].str.isdigit()          # True if all chars are digits
is_alphanumeric_ticket = ~is_numeric_ticket             # True if contains letters or mixed

# Separate DataFrames
numeric_tickets_df = df[is_numeric_ticket]
alphanumeric_tickets_df = df[is_alphanumeric_ticket]

# Check counts
print(f"Numeric tickets count: {len(numeric_tickets_df)}")
print(f"Alphanumeric tickets count: {len(alphanumeric_tickets_df)}")
```

```python
print("Survival rates for numeric ticket holders:")
print(numeric_tickets_df['Survived'].value_counts(normalize=True) * 100)

print("\nSurvival rates for alphanumeric ticket holders:")
print(alphanumeric_tickets_df['Survived'].value_counts(normalize=True) * 100)
```

```python
# 1. Identify ticket type
df['TicketType'] = df['Ticket'].str.isdigit().map({True: 'Numeric', False: 'Alphanumeric'})
survival_analysis = df.groupby(['TicketType', 'Deck', 'AgeGroup'])['Survived'].agg(
    count='count',
    survival_rate='mean'
).reset_index()

# 5. Convert survival_rate to percentage and round
survival_analysis['survival_rate'] = (survival_analysis['survival_rate'] * 100).round(2)

# 6. Show the result
survival_analysis.sort_values('survival_rate',ascending = False)
```

# Which group values (engineered) have the highest internal survival variance across members with different AgeGroup and Sex?

```python
# Group by TicketType, Deck, AgeGroup, and Sex
group_variance = df.dropna(subset=['AgeGroup', 'Sex', 'Deck']).groupby(
    ['TicketType', 'Deck', 'AgeGroup', 'Sex']
)['Survived'].agg(
    count='count',
    survival_variance='var'
).reset_index()

# Drop NaN variance (e.g., when only one passenger in a group)
group_variance = group_variance.dropna(subset=['survival_variance'])

# Sort by highest variance
group_variance = group_variance.sort_values(by='survival_variance', ascending=False).reset_index(drop=True)

# Optional: Round for readability
group_variance['survival_variance'] = group_variance['survival_variance'].round(4)

# Display
print(group_variance.head(10))  # Top 10 groups with highest survival variance
```

```python
# Highest survival variance (0.50) appears in small groups (count = 2), indicating a perfect split—one survived, one died.

# These high-variance groups include both numeric and alphanumeric ticket types, across Decks B, C, D, E and mostly adult males/females.

# Slightly lower variance (0.33) appears in larger groups (count = 3), still reflecting mixed survival outcomes.

# Groups with 5 members (variance around 0.30) suggest moderate inconsistency—some passengers survived, some didn’t.

# Overall, greater group size tends to reduce variance, but even among larger male groups (Deck A, B, C, D), survival outcomes weren’t consistent.
```
